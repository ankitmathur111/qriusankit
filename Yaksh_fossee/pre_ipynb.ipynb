```python
import numpy as np
import matplotlib.pyplot as plt

x=np.linspace(0,2*np.pi,50)
plt.plot(x,np.sin(x))
```

```python
x=np.linspace(-5*np.pi, 5*np.pi, 500)
plt.plot(x,x,'b')
plt.plot(x,-x,'b')
plt.plot(x,np.sin(x),'g')
plt.plot(x,x*np.sin(x),'r')
plt.legend(['x','-x','sin(x)','x sin(x)'], loc='upper right')
plt.annotate('origin',[0,0])
plt.xlim(-5*np.pi, 5*np.pi)
plt.ylim(-5*np.pi, 5*np.pi)
```
